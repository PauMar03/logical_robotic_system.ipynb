```python
# PUT YOUR CODE HERE: Compute the lines.
import numpy as np
import requests
from io import BytesIO
from PIL import Image
import matplotlib.pyplot as plt

# URL de la imagen en el repositorio de GitHub
url = "https://raw.githubusercontent.com/fquinterov/computer_vision_course/main/resources/images/train_tracks.png"

# Cargar la imagen
response = requests.get(url)
img = Image.open(BytesIO(response.content))

# Definir los puntos en coordenadas cartesianas
p1 = np.array([390, 200, 1])  # Homogeneous coordinates (x, y, 1)
p2 = np.array([280, 400, 1])
p3 = np.array([520, 200, 1])
p4 = np.array([620, 400, 1])

# Graficar la imagen
plt.imshow(img)
plt.scatter([p1[0], p2[0], p3[0], p4[0]], [p1[1], p2[1], p3[1], p4[1]], c='red', marker='o', label='Puntos')

# Calcular las líneas en coordenadas homogéneas
l1 = np.cross(p1, p2)  # Línea que pasa por p1 y p2
l2 = np.cross(p3, p4)  # Línea que pasa por p3 y p4

# Función para graficar líneas a partir de su ecuación en coordenadas homogéneas
def linePoints(line, ref=[-1., 1.]):
    """Dado a, b, c para la ecuación ax + by + c = 0,
    devuelve y grafica un par de puntos basados en valores de referencia."""
    a, b, c = line
    if a == 0 and b == 0:
        raise Exception("linePoints: a y b no pueden ser ambos cero")
    else:
        points = np.array([[-c/a, p] if b == 0 else [p, (-c - a*p)/b] for p in ref])
        plt.plot(points[:, 0], points[:, 1], linestyle='dashed', label='Línea')

# Graficar las líneas sobre la imagen
linePoints(l1, [220, 500])
linePoints(l2, [420, 680])

plt.legend()
plt.show()
```

```python
#PUT YOUR CODE HERE:
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from skimage import io, util, filters
from scipy.ndimage import gaussian_filter
from skimage.filters import median
from skimage.morphology import disk

# URL de la imagen en el sitio web de UTB
url = "https://www.utb.edu.co/wp-content/uploads/2024/04/background-unidades-de-apoyo-de-rectoria-1.png"

# Cargar la imagen y convertirla a escala de grises
image = io.imread(url)
image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

# Mostrar la imagen original en escala de grises
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Imagen Original")
plt.axis("off")

# Agregar ruido de sal y pimienta
image_sp = util.random_noise(image, mode='s&p')

# Mostrar la imagen con ruido
plt.subplot(1, 3, 2)
plt.imshow(image_sp, cmap='gray')
plt.title("Imagen con Ruido (Sal y Pimienta)")
plt.axis("off")

# Aplicar filtro para eliminar ruido (usaremos filtro mediano)
image_filtered = median(image_sp, disk(3))  # Se usa un disco de radio 3 para el filtro mediano

# Mostrar la imagen filtrada
plt.subplot(1, 3, 3)
plt.imshow(image_filtered, cmap='gray')
plt.title("Imagen Filtrada (Mediana)")
plt.axis("off")

plt.show()
```
